SECTION 1 - Calculate wavefunction $|\phi^{(1)}(t + \delta t)>$ obtained by evolving the wavefunction with the non-hermitian Hamiltonian

$$H = H_{s} - \frac{i\hbar}{2}\sum_{m}C^{\dagger}_{m}C_{m}.$$

For sufficiently small $\delta t$, this gives:

$$|\phi^{(1)}(t + \delta t)> = \left( 1 - \frac{iH\delta t}{\hbar}\right)|\phi(t)> .$$ 


Need to normalise the wavefunciton:

$$<\phi^{(1)}(t + \delta t)|\phi^{(1)}(t + \delta t)> =<\phi(t)| \left( 1 + \frac{iH\delta t}{\hbar} \right)\left( 1 - \frac{iH\delta t}{\hbar}\right)|\phi(t)> = 1 - \delta p , $$

$$\delta p = \delta t \frac{i}{\hbar} <\phi(t)| H - H^{\dagger} |\phi (t)> = \sum_{m} \delta p_{m},$$

$$ \delta p_{m} = \delta t <\phi (t)|C^{\dagger}_{m} C_{m}| \phi (t)> .$$



Pick a random number $e$.
If $\delta p < e$ then no quantum jump occurs.
The state is then given by:

$$|\phi(t + \delta t)> = \frac{|\phi^{(1)}(t + \delta t)>}{(1 - \delta p)^{\frac{1}{2}}}$$

If $\delta p > e$ then a quantum jump occurs and we choose a new normalised wavefunction among the different states $C_{m}|\phi (t)>$, according to the probability law $\prod_{m}= \frac{\delta p_{m}}{\delta p}$. The state is then given by:

$$|\phi(t + \delta t)> = \frac{C_{m}|\phi (t)>}{(\delta p_{m} / \delta t)^{\frac{1}{2}}}$$


For spontaneous emission of a two level system, there is just one operator $C$, given by:

$$ C = \sqrt{\Gamma} \sigma^{-} $$

$$\sigma^{-} = b$$

Therefore when a jump occurs we will transition from state $|1>$ to state $|0>$

Thus, we just need the form of the wavefunction if the quantum jump does not occur.

For this system, $H_{s} = \hbar \omega _{0}b^{\dagger} b$

Thus, 
$$ H = \hbar \omega _{0}b^{\dagger} b - \frac{i\hbar}{2}\Gamma b^{\dagger} b $$

$$ b^{\dagger} b = N,$$ 
where N is the number operator, thus, 
$$ H = \left(\omega _{0} - \frac{i}{2}\Gamma \right) \hbar N $$

We start with an inital wavefunction given by

$$|\phi(0)> = \alpha_{0}|0> + \beta_{0}|1>$$ 

Evolving the wavefuntion through time using $U(t) = e^{-iHt/\hbar}$ we find:

$$|\phi^{(1)}(t + \delta t)>  = \alpha_{0}|0> + \beta_{0}exp(-i\omega_{0} \delta t)exp(-\Gamma \delta t/2)|1>$$.

$\delta p$ is given by $\Gamma|\beta|^{2}\delta t$ .

Assuming $\delta t$ is sufficiently small, the normalised wavefucntion if no quantum jump occurs is given by,

$$|\phi(\delta t)> = \alpha_{0} \left (1 + \frac{\Gamma \delta t}{2} |\beta_{0}|^{2} \right)|0> + \beta_{0} \left (1 - \frac{\Gamma \delta t}{2}|\alpha _{0}|^{2} \right ) exp(-i\omega_{0} \delta t)|1>$$

```python
def dp(wf, gamma, dt):
    return gamma*np.abs(wf[1])**2*dt
    
def newa(wf, gamma, dt):
    return wf[0]*(1+((gamma*dt)/2)*(np.abs(wf[1])**2))

def newb(wf, gamma, dt):
    return wf[1]*(1-((gamma*dt)/2)*(np.abs(wf[0])**2))#*np.exp(-1j*omega0*dt)
    
    
def newwf(wf, gamma, dt):
    return [newa(wf,gamma,dt),newb(wf,gamma,dt)]

def expdecay(gamma,dt):
    return np.abs(np.exp(-gamma*dt*(1/2)))**2
```

```python
#setting constants
t_max=10
t_steps=1000

dt=t_max/t_steps
gamma=1.0

N=100

b0=1/np.sqrt(2)
a0=np.sqrt(1-b0**2)


#Plotting code lifted straight from Kalus OBE codes
fig = plt.figure(1)

fs=20

plt.rcParams.update({'font.size': fs})
plt.rcParams.update({'figure.figsize':(10,10)})

lw=6
ax_lw=2
tick_lw=3
tick_len=10


ax=fig.add_axes([0.0, 0.0, 1.0, 1.0])

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(ax_lw)
ax.xaxis.set_ticks_position('both')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_tick_params(width=ax_lw,length=tick_len,direction='in')
ax.yaxis.set_tick_params(width=ax_lw,length=tick_len,direction='in')

color=iter(cm.rainbow(np.linspace(0,1,N)))
total=np.zeros(t_steps)
#######


jump = []
#main code which runs over N particles
b0=1/np.sqrt(2)
a0=np.sqrt(1-b0**2)

for nn in range(0,N):
    ts = [] #empty list of times
    bs = [] #empty list of excited state probabilites
 
    wf=[a0, b0]
    
    for tt in range(0,t_steps+1):

        nwf=newwf(wf, gamma, dt) #calculates new wavefunction at time+dt
        p = dp(wf,gamma,dt)
        e = np.random.rand(1)[0] #generate a random number between 0 and 1
        
        if p>e:#test for quantum jump
            jump.append(1)
            #print(tt)
            wf = [1,0] #wavefunciton jumps to ground state
        else:
            wf = nwf
            prob_b=np.abs(wf[1])**2  #probability of being in excited state
            bs.append(prob_b)
            ts.append(dt*tt) #time values

    bs=bs[:t_steps]
    ts=ts[:t_steps]
    total+=bs
    
    ax.plot(ts,bs,alpha=0.2,color='gray')
print(len(jump))
ax.plot(ts,bs,alpha=0.2,color='gray',label='trajectories')
ts=np.array(ts)
expys = (b0**2)*expdecay(gamma, ts)


ax.plot(ts,expys,color='blue',label='Exponential Decay') #plots decay of state 1 probability
ax.plot(ts,total/N,color='red',label='Quantum Jump Method') #plots average of trajectories

plt.ylabel('Excited state prob.',fontsize=fs)
plt.xlabel('Time(1/$\Gamma$)')
plt.legend()
plt.show()  
```

```python
#setting constants

N=60

b0=0.9
a0=np.sqrt(1-b0**2)


R=30 #number of runs


#Plotting code lifted straight from Kalus OBE codes
fig = plt.figure(2)

fs=28

plt.rcParams.update({'font.size': fs})
plt.rcParams.update({'figure.figsize':(10,5)})

lw=6
ax_lw=2
tick_lw=3
tick_len=10


ax=fig.add_axes([0.0, 0.0, 1.0, 1.0])

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(ax_lw)
ax.xaxis.set_ticks_position('both')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_tick_params(width=ax_lw,length=tick_len,direction='in')
ax.yaxis.set_tick_params(width=ax_lw,length=tick_len,direction='in')

color=iter(cm.rainbow(np.linspace(0,1,R)))
#######



for rr in range(0,R):
    
    total=np.zeros(t_steps)#used to calculate the average trajectory of N particles
    for nn in range(0,N):
        ts = [] #empty list of times
        bs = [] #empty list of excited state probabilites

        wf=[a0, b0]

        for tt in range(0,t_steps+1):

            nwf=newwf(wf, gamma, dt) #calculates new wavefunction at time+dt
            p = dp(wf,gamma,dt)
            e = np.random.rand(1)[0] #generate a random number between 0 and 1

            if p>e:#test for quantum jump
                #print('Quantum Jump!')
                #print(tt)
                wf = [1,0] #wavefunciton jumps to ground state
            else:
                wf = nwf
                prob_b=np.abs(wf[1])**2  #probability of being in excited state
                bs.append(prob_b)
                ts.append(dt*tt) #time values

        bs=bs[:t_steps]
        ts=ts[:t_steps]
        total+=bs

    c=next(color)
    ax.plot(ts,total/N,color=c,alpha=.8) #plots average of N trajectories

ts=np.array(ts)
expys = (b0**2)*expdecay(gamma, ts)

ax.plot(ts,expys,color='black',label='analytical decay') #plots analytical decay of the excited state probability

plt.legend()
plt.ylabel('Excited state prob.',fontsize=fs)

plt.show()  
```
